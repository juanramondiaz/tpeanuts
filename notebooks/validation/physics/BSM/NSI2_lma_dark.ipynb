## 0. Theory Background

### 0.1 Standard MSW Oscillations — The LMA Solution

In the two-flavour approximation relevant for solar neutrinos, the propagation
Hamiltonian in the flavour basis is *(Wolfenstein 1978; Mikheyev & Smirnov 1985)*

$$H = \frac{\Delta m^2_{21}}{4E}
\begin{pmatrix}-\cos 2\theta_{12} & \sin 2\theta_{12}\\
\sin 2\theta_{12} & \cos 2\theta_{12}\end{pmatrix}
+ \frac{V_e}{2}\begin{pmatrix}1 & 0\\ 0 & -1\end{pmatrix},
\qquad V_e = \sqrt{2}\,G_F\,n_e.$$

The **MSW resonance** occurs at $V_e^{\rm res} = \Delta m^2_{21}\cos 2\theta_{12}/(2E)$.
For the LMA solution ($\theta_{12}\approx33.4°$, first octant) this requires positive
density — always satisfied inside the Sun. Adiabatic propagation projects $\nu_e$ onto
the heavier matter eigenstate, giving

$$P_{ee}^{\rm SM}\approx\sin^2\theta_{12}\approx 0.30 \quad (\text{high }E),
\qquad P_{ee}^{\rm SM}\approx 0.55 \quad (\text{low }E, \text{vacuum-averaged}).$$

---

### 0.2 NSI and the Effective Matter Potential

Non-Standard Interactions (NSI) modify the neutrino forward-scattering amplitude
through dimensionless couplings $\varepsilon_{\alpha\beta}$ *(Wolfenstein 1978;
Guzzo et al. 1991)*. The diagonal electron-neutrino coupling $\varepsilon_{ee}$
rescales the CC potential:

$$V_e^{\rm eff} = (1 + \varepsilon_{ee})\,V_e.$$

For $\varepsilon_{ee} = -2$ the potential **reverses sign**: $V_e^{\rm eff} = -V_e$.
The resonance condition then requires negative density — unphysical — so no MSW
resonance occurs inside the Sun for standard $\theta_{12} < 45°$.

---

### 0.3 The Discrete LMA-Dark Degeneracy

With $V_e^{\rm eff}<0$ and $\theta_{12}' = \pi/2 - \theta_{12}^{\rm SM}\approx56.6°$
(second octant), the combination *(Esteban et al. 2018)*

$$\varepsilon_{ee} = -2,\quad \theta_{12}' = \pi/2 - \theta_{12}^{\rm SM},$$

yields the same high-energy survival probability as the SM LMA solution:

$$P_{ee}^{\rm LMA\text{-}Dark}\approx\cos^2\theta_{12}' = \sin^2\theta_{12}^{\rm SM}\approx 0.30.$$

At low energy both solutions give the same vacuum-averaged
$P_{ee}\approx1-\frac{1}{2}\sin^2 2\theta_{12}\approx0.55$, because
$\sin^2 2\theta' = \sin^2(\pi - 2\theta) = \sin^2 2\theta$.
The two solutions are therefore **exactly degenerate** in all daytime solar data.

---

### 0.4 Observable Discriminator — Day–Night Asymmetry

Earth matter regeneration breaks the degeneracy. The day–night asymmetry is

$$A_{\rm DN} = 2\,\frac{P_{ee}^{\rm night} - P_{ee}^{\rm day}}{P_{ee}^{\rm night} + P_{ee}^{\rm day}}.$$

- **SM LMA** ($\varepsilon_{ee}=0$): Earth potential is positive → regenerates $\nu_e$
  → $A_{\rm DN}>0$.
- **LMA-Dark** ($\varepsilon_{ee}=-2$): Earth potential acquires effective negative sign
  → depletes $\nu_e$ → $A_{\rm DN}<0$.

This sign reversal is the cleanest experimental discriminator
*(Minakata & Nunokawa 1997)*.

---

### 0.5 Experimental Status

| Observable | LMA | LMA-Dark | Measurement |
|------------|-----|----------|-------------|
| $P_{ee}$ (low $E$) | $\approx0.55$ | $\approx0.55$ | $0.57\pm0.09$ (Borexino $pp$) |
| $P_{ee}$ (high $E$) | $\approx0.30$ | $\approx0.30$ | $0.317\pm0.016$ (SNO) |
| $A_{\rm DN}$ ($^8$B, $>5.5$ MeV) | $>0$ | $<0$ | $+3.3\pm1.0\%$ (Super-K) |

The Super-K day–night measurement disfavours LMA-Dark at $\sim3\sigma$.
A global NSI fit places the LMA-Dark minimum at $\Delta\chi^2\approx10$–20 above LMA.

### 2.3 Helper Functions

```python
VE_PER_MOLCM3 = 7.63e-14  # eV per (mol/cm^3) = sqrt(2)*G_F*N_A


def msw_eigenvalues(theta, Ve_arr, Dm21_eV2, E_eV, eps_ee=0.0):
    a    = Dm21_eV2 / (4.0 * E_eV)
    Veff = (1.0 + eps_ee) * Ve_arr
    H11  = -a * np.cos(2 * theta) + Veff / 2
    H12  =  a * np.sin(2 * theta)
    H22  =  a * np.cos(2 * theta) - Veff / 2
    tr2  = (H11 + H22) / 2
    disc = np.sqrt((H11 - tr2) ** 2 + H12 ** 2)
    return tr2 - disc, tr2 + disc


@torch.no_grad()
def earth_spectrum(mass_weights, oscillation, earth_profile, eta_t, E_grid,
                   depth_m, nsteps=100, epsilon=None):
    P_list = []
    for ei in range(len(E_grid)):
        P_i = pearth_numerical(
            nustate=mass_weights[ei],
            profile_earth=earth_profile,
            oscillation=oscillation,
            E_MeV=E_grid[ei],
            eta=eta_t,
            depth_m=depth_m,
            massbasis=True,
            epsilon=epsilon,
            nsteps=nsteps,
            context=context,
        )
        P_list.append(P_i)
    return torch.stack(P_list, 0).detach().cpu().numpy()
```

## 3. MSW Level Crossing: Hamiltonian Eigenvalues

The eigenvalues of the 2-flavour Hamiltonian
$H = H_{\rm vac}(\theta_{12}, \Delta m^2_{21}, E) + H_{\rm mat}(\varepsilon_{ee})$
plotted as a function of electron density $n_e$ reveal the anti-crossing
(MSW resonance) for SM LMA and its absence for LMA-Dark.

**Expected results:**
- **SM LMA** ($\varepsilon_{ee}=0$, $\theta_{12}=33.4°$): a clear avoided crossing
  appears within the solar-core density band ($n_e\sim0.15$–$0.35$ mol/cm$^3$)
  at $E=10$ MeV, confirming the MSW resonance inside the Sun.
- **LMA-Dark** ($\varepsilon_{ee}=-2$, $\theta_{12}=56.6°$): no anti-crossing at any
  positive density; the levels evolve monotonically without crossing,
  consistent with the resonance being pushed to unphysical (negative) densities.

```python
Dm21_eV2  = float(osc_sm.DeltamSq21)
E_MeV_ref = 10.0
E_eV_ref  = E_MeV_ref * 1e6

ne     = np.logspace(-6, 0.5, 600)
Ve_arr = VE_PER_MOLCM3 * ne

scenarios = [
    (th12_sm,   0.0, "royalblue",
     r"SM LMA ($\theta_{12}=33.4^\circ$, $\varepsilon_{{ee}}=0$)"),
    (th12_dark, -2.0, "firebrick",
     r"LMA-Dark ($\theta_{{12}}=56.6^\circ$, $\varepsilon_{{ee}}=-2$)"),
]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
for ax, (th, eps_ee, color, title) in zip(axes, scenarios):
    lm, lp = msw_eigenvalues(th, Ve_arr, Dm21_eV2, E_eV_ref, eps_ee=eps_ee)
    fac = 1e15
    ax.semilogx(ne, lm * fac, lw=2.5, color=color, label=r"$\lambda_-$ (lighter)")
    ax.semilogx(ne, lp * fac, lw=2.5, color=color, ls="--", label=r"$\lambda_+$ (heavier)")
    ax.text(ne[0] * 1.5, float(lm[0]) * fac + 0.02, r"$\nu_1$ (vac)", fontsize=8)
    ax.text(ne[0] * 1.5, float(lp[0]) * fac + 0.02, r"$\nu_2$ (vac)", fontsize=8)
    ax.axvspan(0.15, 0.35, alpha=0.08, color="gold", label="solar core range")
    ax.axvline(0.25, color="goldenrod", lw=1.0, ls=":")
    ax.set_xlabel(r"$n_e$ [mol/cm$^3$]")
    ax.set_ylabel(r"Eigenvalue [$\times 10^{{-15}}$ eV]")
    ax.set_title(title, fontsize=11)
    ax.legend(fontsize=9)
    ax.grid(True, which="both", alpha=0.3)

fig.suptitle(
    f"2-Flavour MSW eigenvalues  (E = {E_MeV_ref:.0f} MeV,  "
    r"$\Delta m^2_{{21}}$" + f" = {Dm21_eV2:.2e} eV$^2$)",
    fontsize=12,
)
fig.tight_layout()
save_and_show("bsm3_fig3_eigenvalues.png", fig, output_dir=OUTPUT_DIR,
              show_plots=config.show_plots)
print("Left : anti-crossing visible inside solar-core band (MSW resonance).")
print("Right: no anti-crossing — LMA-Dark avoids the solar resonance in normal matter.")
```

## 4. Daytime Solar Survival Probability $P_{ee}(E)$

The adiabatic survival probability integrated over the $^8$B solar production
profile is computed for two scenarios accessible in the tpeanuts daytime pipeline:

- **SM** ($\theta_{12}=33.41°$, no NSI): standard LMA prediction.
- **Dark-angular** ($\theta_{12}=56.59°$, no NSI in Sun): second-octant rotation
  *without* the compensating NSI in the solar medium — not the true degenerate solution.
- **True LMA-Dark** (analytic): exact degenerate prediction
  $P_{ee}\approx\cos^2\theta_{12}^{\rm Dark}=\sin^2\theta_{12}^{\rm SM}$,
  derived analytically and shown as a horizontal dashed line.

**Expected results:**
- **SM (LMA)**: $P_{ee}$ falls from $\approx0.55$ at low $E$ to $\approx0.30$ at
  $E\gtrsim8$ MeV (MSW adiabatic suppression).
- **Dark-angular only** (no NSI in Sun): rises toward $\sin^2(56.6°)\approx0.70$
  at high $E$ — opposite to SM because $\theta_{12}>45°$ enhances the positive
  MSW potential instead of suppressing it.
- **True LMA-Dark** (analytic): identical to SM at all energies — the degeneracy.

```python
solar_profile = SolarProfile.default(context=context)
print(solar_profile)

SOURCE = "8B"
E_MeV  = torch.linspace(0.5, 18.0, 100, dtype=context.dtype, device=context.device)

with torch.no_grad():
    P_sm   = psolar(osc_sm,   E_MeV, solar_profile, SOURCE)
    P_dark = psolar(osc_dark, E_MeV, solar_profile, SOURCE)

Pee_sm   = P_sm[:,   0].cpu().numpy()
Pee_dark = P_dark[:, 0].cpu().numpy()
E_np     = E_MeV.cpu().numpy()

Pee_vac  = math.cos(th12_sm)**4 + math.sin(th12_sm)**4
Pee_lmad = math.cos(th12_dark)**2

print(f"SM   P_ee at 10 MeV : {float(np.interp(10.0, E_np, Pee_sm)):.3f}")
print(f"Dark P_ee at 10 MeV : {float(np.interp(10.0, E_np, Pee_dark)):.3f}")
print(f"True LMA-Dark (adiabatic, analytic) : {Pee_lmad:.3f}")
print(f"Vacuum average P_ee (both)          : {Pee_vac:.3f}")
```

```python
data = {
    r"$pp$  Borexino":   (0.26,  0.57, 0.08, 0.09),
    r"$^7$Be  Borexino": (0.86,  0.53, 0.05, 0.05),
    r"pep  Borexino":    (1.44,  0.43, 0.10, 0.10),
    r"$^8$B  SNO":       (9.0,   0.317, 0.016, 0.016),
}

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(E_np, Pee_sm,   color="royalblue",  lw=2.5,
        label=r"SM LMA ($\theta_{{12}}=33.4^\circ$, no NSI)")
ax.plot(E_np, Pee_dark, color="darkorange", lw=2.5, ls="--",
        label=r"Dark-angular ($\theta_{{12}}=56.6^\circ$, no NSI in $\odot$) — not degenerate")
ax.axhline(Pee_lmad, color="firebrick", lw=1.5, ls=":",
           label=rf"True LMA-Dark (analytic adiabatic, $\approx{Pee_lmad:.2f}$)")
ax.axhline(Pee_vac, color="gray", lw=1.0, ls=":", alpha=0.5)
ax.text(17.5, Pee_vac + 0.01, f"{Pee_vac:.2f} (vacuum avg)", color="gray",
        ha="right", fontsize=8)

for label, (E0, P0, el, eu) in data.items():
    col = "crimson" if "SNO" in label else "forestgreen"
    ax.errorbar(E0, P0, yerr=[[el], [eu]], fmt="o", color=col, ms=7, capsize=4, zorder=5)
    ax.annotate(label, (E0, P0), xytext=(E0 + 0.2, P0 + 0.035), fontsize=8.5, color=col)

ax.set_xlabel(r"$E_\nu$ [MeV]")
ax.set_ylabel(r"$P_{{ee}}$ (daytime)")
ax.set_title(r"Solar $\nu_e$ daytime survival probability $P_{{ee}}(E)$")
ax.set_xlim(0, 19)
ax.set_ylim(0.15, 0.82)
ax.legend(fontsize=9, loc="lower right")
ax.grid(True, which="both", alpha=0.3)
fig.tight_layout()
save_and_show("bsm3_fig4_pee_daytime.png", fig, output_dir=OUTPUT_DIR,
              show_plots=config.show_plots)
```

## 5. Day–Night Asymmetry $A_{\rm DN}$: The Key Discriminator

The Earth regeneration probability modifies the $\nu_e$ fraction during nighttime
traversal. Three scenarios are computed with `pearth_numerical` (NSI supported):

1. **SM solar weights + no Earth NSI** — standard LMA Earth regeneration.
2. **SM solar weights + Earth NSI** ($\varepsilon_{ee}=-2$) — illustrates the sign
   reversal from Earth NSI alone.
3. **Dark-angular solar weights + Earth NSI** ($\varepsilon_{ee}=-2$) — closest
   available approximation to the full LMA-Dark nighttime prediction.

**Expected results:**
Super-K measures $A_{\rm DN}=+3.3\pm1.0\%$ for $^8$B above 5.5 MeV (arXiv:1606.07011):
- **Scenario 1** (SM LMA): $A_{\rm DN}\approx+1$–3%, consistent with Super-K.
- **Scenario 2** (SM + NSI Earth): $A_{\rm DN}$ flips negative — sign reversal from
  the NSI Earth potential.
- **Scenario 3** (LMA-Dark approx): $A_{\rm DN}<0$, disfavoured by Super-K at $\sim3\sigma$.

```python
DEPTH_M   = 1.0e3   # 1 km underground
ETA_DEG   = 30.0    # nadir angle (typical Super-K geometry)
NSTEPS    = 100

eta_t = torch.tensor(math.radians(ETA_DEG), dtype=context.dtype, device=context.device)

earth_profile = EarthProfile(
    params=EarthParameters(
        method="numerical", depth_m=DEPTH_M, nsteps=NSTEPS,
        profile_perturbative_name='even_power',
        profile_perturbative_kwargs={'density_file': str(EARTH_DENSITY_CSV),
                                     'tabulated_density': False},
    ),
    context=context,
)

print(earth_profile)
print(f"Nadir angle : {ETA_DEG:.1f}°   depth : {DEPTH_M/1e3:.0f} km")
```

```python
SOURCE_E = "8B"
E_earth  = torch.linspace(4.0, 16.0, 40, dtype=context.dtype, device=context.device)

with torch.no_grad():
    w_sm   = solar_probability_mass(osc_sm,   E_earth, solar_profile, SOURCE_E)
    w_dark = solar_probability_mass(osc_dark, E_earth, solar_profile, SOURCE_E)

idx10 = int(np.argmin(np.abs(E_earth.cpu().numpy() - 10.0)))
print(f"Solar mass weights at E~10 MeV:")
print(f"  SM   w = {w_sm[idx10].tolist()}")
print(f"  Dark w = {w_dark[idx10].tolist()}")
```

```python
eps_dark_t = nsi_dark.epsilon_tensor(context.device, context.dtype)

print("Computing nighttime probabilities ...")
print("  (1) SM solar + no Earth NSI ...", end=" ", flush=True)
P_night_sm_no  = earth_spectrum(w_sm,   osc_sm,   earth_profile, eta_t, E_earth,
                                DEPTH_M, NSTEPS, epsilon=None)
print("done")
print("  (2) SM solar + eps_ee=-2 Earth ...", end=" ", flush=True)
P_night_sm_nsi = earth_spectrum(w_sm,   osc_sm,   earth_profile, eta_t, E_earth,
                                DEPTH_M, NSTEPS, epsilon=eps_dark_t)
print("done")
print("  (3) Dark solar + eps_ee=-2 Earth ...", end=" ", flush=True)
P_night_dk_nsi = earth_spectrum(w_dark, osc_dark, earth_profile, eta_t, E_earth,
                                DEPTH_M, NSTEPS, epsilon=eps_dark_t)
print("done")

with torch.no_grad():
    Pday_sm   = psolar(osc_sm,   E_earth, solar_profile, SOURCE_E)[:, 0].cpu().numpy()
    Pday_dark = psolar(osc_dark, E_earth, solar_profile, SOURCE_E)[:, 0].cpu().numpy()

E_earth_np = E_earth.cpu().numpy()

def adn(Pnight_ee, Pday_ee):
    return 2.0 * (Pnight_ee - Pday_ee) / (Pnight_ee + Pday_ee)

A1 = adn(P_night_sm_no[:, 0],  Pday_sm)
A2 = adn(P_night_sm_nsi[:, 0], Pday_sm)
A3 = adn(P_night_dk_nsi[:, 0], Pday_dark)

print()
print(f"A_DN at E=10 MeV:")
print(f"  (1) SM + no Earth NSI:      {float(np.interp(10.0, E_earth_np, A1))*100:+.2f}%")
print(f"  (2) SM + eps_ee=-2 Earth:   {float(np.interp(10.0, E_earth_np, A2))*100:+.2f}%")
print(f"  (3) Dark + eps_ee=-2 Earth: {float(np.interp(10.0, E_earth_np, A3))*100:+.2f}%")
print("Super-K (arXiv:1606.07011): A_DN = +3.3 +/- 1.0%")
```

```python
fig, ax = plt.subplots(figsize=(10, 5.5))

ax.plot(E_earth_np, A1 * 100, color="royalblue", lw=2.5,
        label=r"SM LMA ($\varepsilon_{{ee}}=0$)")
ax.plot(E_earth_np, A2 * 100, color="royalblue", lw=2.0, ls=":",
        label=r"SM solar + $\varepsilon_{{ee}}=-2$ Earth  [sign-flip demo]")
ax.plot(E_earth_np, A3 * 100, color="firebrick", lw=2.5, ls="--",
        label=r"LMA-Dark approx. ($\theta_{{12}}=56.6^\circ$, $\varepsilon_{{ee}}=-2$ Earth)")

sk_val, sk_err = 3.3, 1.0
ax.axhspan(sk_val - sk_err, sk_val + sk_err, alpha=0.15, color="forestgreen")
ax.axhline(sk_val, color="forestgreen", lw=1.5,
           label=rf"Super-K: $A_{{\rm DN}}={sk_val:.1f}\pm{sk_err:.1f}\%$ (arXiv:1606.07011)")
ax.axhline(0.0, color="black", lw=0.8)

ax.set_xlabel(r"$E_\nu$ [MeV]")
ax.set_ylabel(r"$A_{{\rm DN}}$ [%]")
ax.set_title(
    rf"Day-Night asymmetry  (nadir $\eta={ETA_DEG:.0f}^\circ$, depth = {DEPTH_M/1e3:.0f} km)"
)
ax.set_xlim(4, 16)
ax.legend(fontsize=9, loc="lower right")
ax.grid(True, which="both", alpha=0.3)
fig.tight_layout()
save_and_show("bsm3_fig5_adn.png", fig, output_dir=OUTPUT_DIR,
              show_plots=config.show_plots)

print("A_DN > 0  ->  Earth regenerates nu_e  ->  LMA preferred")
print("A_DN < 0  ->  Earth depletes nu_e     ->  LMA-Dark signature")
```
