# Solar Neutrino Analysis 3 — Adiabatic Approximation
---

This notebook quantifies **how adiabatic the MSW resonance is** in the solar interior. The central diagnostic is the dimensionless adiabaticity parameter $\gamma_{\rm res}(E)$: 
- When $\gamma_{\rm res}\gg 1$ the neutrino stays on its local matter eigenstate throughout propagation (perfect adiabaticity);
- When $\gamma_{\rm res}\sim 1$ a Landau-Zener jump becomes probable (treated in Next Notebook).

## 0. Theory Background

### 0.1 The Adiabatic Condition

A quantum state evolves adiabatically when the external parameters change slowly compared with the inverse of the energy-level gap. For solar MSW propagation, the relevant condition compares the **rate of change of the mixing angle** to the **local oscillation frequency** *(Giunti & Kim 2007)*:

$$\left|\frac{\dot{\theta}_{12}^M}{\omega}\right| \ll 1, \qquad \omega = \frac{\Delta m^2_{21}}{2E\hbar c}\sqrt{(\cos 2\theta_{12}-V_k)^2+\sin^2 2\theta_{12}},$$

where the dot denotes a derivative with respect to the propagation coordinate $r$. This ratio is largest at the **resonance** ($V_k = \cos 2\theta_{12}$, gap minimal), so adiabaticity is most likely to fail there.

---

### 0.2 Density Scale Height $L_n$

The rate at which the mixing angle changes is governed by how fast the electron density varies. The **density scale height** quantifies this:

$$L_n(r) = \left|\frac{n_e}{\mathrm{d}n_e/\mathrm{d}r}\right|,$$

measured in the same units as $r$ (metres when $r$ is expressed in metres). A **large** $L_n$ means density changes slowly — a necessary (though not sufficient) condition for adiabaticity. Near the solar core $L_n$ is small ($\sim 10^4$ km); in the outer envelope it grows considerably.

---

### 0.3 Adiabaticity Parameter at the Resonance

Evaluating the adiabatic ratio at the resonance and using a locally exponential density profile yields the **Parke adiabaticity parameter** *(Parke 1986)*:

$$\boxed{\gamma_{\rm res}(E) = \frac{\Delta m^2_{21}\,\sin^2 2\theta_{12}}{2E\,\hbar c\,\cos 2\theta_{12}}\,L_n(r_{\rm res}(E))},$$

where:
- $L_n(r_{\rm res})$ is the scale height **at the resonance radius**, in metres.
- The factor $\sin^2 2\theta_{12}/\cos 2\theta_{12}$ encodes the mixing angle dependence.
- $\gamma_{\rm res}\propto \Delta m^2_{21}/E$: smaller splitting or higher energy → less adiabatic.

For standard LMA at $E=10$ MeV: $\gamma_{\rm res}\approx 1400\gg 1$ — the adiabatic approximation is **excellent**.

---

### 0.4 Critical Energy $E_{\rm crit}$

The energy at which $\gamma_{\rm res}(E_{\rm crit}) = 1$ marks the onset of significant non-adiabatic effects. Below $E_{\rm crit}$ the propagation is safely adiabatic; above it a Landau-Zener correction is needed.

For LMA parameters $E_{\rm crit}$ falls far below the pp threshold ($\ll 0.1$ MeV), confirming that the adiabatic approximation is valid for all observable solar sources. For hypothetical scenarios with smaller $\Delta m^2_{21}$, $E_{\rm crit}$ rises into the observable range.

---

### 2.2 Configuration

Three oscillation scenarios are used to explore different adiabaticity regimes:

| Label | $\Delta m^2_{21}$ [eV²] | $\gamma_{\rm res}$ at 10 MeV | Regime |
|-------|------------------------|-------------------------------|--------|
| **LMA** | $7.41\times10^{-5}$ (NuFIT 5.2) | $\sim 2000$ | Strongly adiabatic |
| **Hyp** | $7.41\times10^{-7}$ (÷100) | $\sim 30$ | Well adiabatic |
| **Non Adiabatic (N.A.)** | $7.41\times10^{-8}$ (÷1000) | $\sim 1.5$ | Onset of LZ jumps |

All three use the same solar profile (zenodo SF3-AGSS09, $r\in[0,1]\,R_\odot$) since the hypothetical resonances sit in the outer envelope.

```python
osc_lma      = PropagationConfig.oscillation_parameters_from_preset(context=ctx, antinu=False)
osc_hyp      = dataclasses.replace(osc_lma, mass_spectrum=dataclasses.replace(osc_lma.mass_spectrum, DeltamSq21=osc_lma.mass_spectrum.DeltamSq21 * 0.01))
osc_nonadiab = dataclasses.replace(osc_lma, mass_spectrum=dataclasses.replace(osc_lma.mass_spectrum, DeltamSq21=osc_lma.mass_spectrum.DeltamSq21 * 0.001))

profile  = SolarProfile.default(context=ctx)
RHO_GRID = profile.radius
NE_GRID  = profile.density

THETA12 = float(osc_lma.pmns.params.theta12.detach().cpu())
DM21    = float(osc_lma.mass_spectrum.DeltamSq21.detach().cpu())

PROBE_ENERGIES = [0.5, 2.0, 5.0, 10.0, 15.0]

print(f'theta12 = {THETA12*180/math.pi:.2f} deg')
print(f'Dm21    = {DM21:.3e} eV²')
print(f'\n{profile}')
```

### 2.3 Helpers

`_gamma_res_grid` evaluates $\gamma_{\rm res}(E)$ for a given scenario by interpolating the density and its gradient to the resonance radius, then applying the Parke formula. Returns NaN where no resonance exists.

```python
def _gamma_res_grid(osc, E_grid, prof):
    """Compute gamma_res(E) for a given oscillation scenario and solar profile."""
    th12   = osc.pmns.params.theta12
    dm21   = osc.mass_spectrum.DeltamSq21
    sin2   = torch.sin(2.0 * th12)
    cos2   = torch.cos(2.0 * th12)
    kw     = dict(device=prof.radius.device, dtype=prof.radius.dtype)
    dne_dr = density_gradient(prof)
    r_res  = resonance_radius(osc, E_grid, prof)          # (n_E,), NaN if absent
    has    = torch.isfinite(r_res)
    r_safe = torch.where(has, r_res,
                         prof.radius[prof.radius.shape[0]//2].expand_as(r_res))
    ne_r   = interp1d_linear(r_safe, prof.radius, prof.density, **kw)
    dne_r  = interp1d_linear(r_safe, prof.radius, dne_dr,       **kw)
    tiny   = torch.finfo(prof.radius.dtype).tiny
    L_n_m  = (ne_r.abs() / (dne_r.abs() + tiny)) * constant.R_SUN
    hbarc  = constant.HBARC_MeV_m * 1e6            # eV·m
    kin    = dm21 / (2.0 * E_grid * 1e6 * hbarc)   # 1/m
    gamma  = kin * (sin2**2 / cos2) * L_n_m
    return torch.where(has, gamma, torch.full_like(gamma, float('nan')))

print('Helpers defined.')
```

## 3. Density Scale Height $L_n(\rho)$

The density scale height $L_n(r) = |n_e/(\mathrm{d}n_e/\mathrm{d}r)|\times R_\odot$ (in km) tells us how fast the solar density changes at each radius. It directly enters $\gamma_{\rm res}$ through the density gradient at the resonance.

**Expected results:**
- $n_e(r)$ decreases monotonically from the solar centre ($\sim 10^2$ mol/cm³) to near-zero at the surface, consistent with the B16 AGSS09 model.
- $L_n$ is small ($\sim 10^3$–$10^4$ km) in the core where the density profile is steep
- $L_n$ grows to $\sim 10^5$ km in the outer envelope where the density is nearly flat.
- A local dip near $r\approx 0.7\,R_\odot$ (tachocline) may be visible in the gradient.

```python
rho_np  = to_numpy(RHO_GRID)
ne_np   = to_numpy(NE_GRID)
dne_np  = to_numpy(density_gradient(profile))
tiny    = np.finfo(float).tiny

L_n_km  = np.abs(ne_np / (np.abs(dne_np) + tiny)) * constant.R_SUN / 1e3

fig, axes = plt.subplots(1, 2, figsize=(12, 4.8))

axes[0].semilogy(rho_np, np.abs(dne_np), 'C1', lw=2)
axes[0].set_xlabel(r'Solar radius $\rho = r/R_\odot$')
axes[0].set_ylabel(r'$|\mathrm{d}n_e/\mathrm{d}\hat{r}|$ [mol cm$^{-3}$ $R_\odot^{-1}$]')
axes[0].set_title('Electron density gradient magnitude')
axes[0].set_xlim(0, 1)

axes[1].semilogy(rho_np, L_n_km, 'C2', lw=2)
axes[1].set_xlabel(r'Solar radius $\rho = r/R_\odot$')
axes[1].set_ylabel(r'$L_n$ [km]')
axes[1].set_title(r'Density scale height $L_n = |n_e/(\mathrm{d}n_e/\mathrm{d}\hat{r})|\times R_\odot$')
axes[1].set_xlim(0, 1)

fig.suptitle('SF3-AGSS09 — density gradient and scale height', fontweight='bold')
fig.tight_layout()
save_and_show('sn3_fig3_scale_height.png', fig, output_dir=OUTPUT_DIR, show_plots=config.show_plots)

print(f'L_n at core (rho=0.001)   : {L_n_km[0]:.0f} km')
print(f'L_n at rho=0.1            : {float(L_n_km[np.argmin(np.abs(rho_np-0.1))]):.0f} km')
print(f'L_n at rho=0.5            : {float(L_n_km[np.argmin(np.abs(rho_np-0.5))]):.0f} km')
print(f'L_n at rho=0.9            : {float(L_n_km[np.argmin(np.abs(rho_np-0.9))]):.0f} km')
```

## 4. Adiabaticity Map $\gamma(\rho, E)$

The full adiabaticity, for LMA scenario landscape, is shown as a 2-D heatmap over the $(\rho, E)$ plane. At each point $(\rho, E)$ we compute $\gamma$ using the *local* scale height $L_n(\rho)$ instead of evaluating it only at the resonance — this shows the global adiabaticity across the solar volume.

$$\gamma(\rho, E) = K_\odot(E)\,\frac{\sin^2 2\theta_{12}}{\cos 2\theta_{12}}\,\left/\,\left|\frac{\mathrm{d}\ln n_e}{\mathrm{d}\rho}\right|\right.$$

**Expected results:** 
- $\gamma\gg 1$ (yellow/green) across most of the plane for LMA parameters;
- values drop toward $\gamma\sim 1$ (dark) only at very low energies in the outer envelope.
- The $\gamma=1$ contour — if present — marks the non-adiabatic boundary.

```python
ENERGY_GRID = torch.logspace(-1, np.log10(20), 200, dtype=ctx.dtype)

log_ne    = torch.log(NE_GRID.clamp_min(torch.finfo(ctx.dtype).tiny))
d_lnne    = torch.gradient(log_ne, spacing=(RHO_GRID,))[0]   # d(ln ne)/d(rho)
ang_fac   = (torch.sin(2.0 * osc_lma.pmns.params.theta12)**2
             / torch.cos(2.0 * osc_lma.pmns.params.theta12))

# gamma(rho, E) over the full grid
gamma_map = torch.zeros(len(ENERGY_GRID), len(RHO_GRID), dtype=ctx.dtype)
for i, E_val in enumerate(ENERGY_GRID):
    K_sun = kinetic_potential(osc_lma.mass_spectrum.DeltamSq21, E_val, evolution_scale_m=constant.R_SUN)
    gamma_map[i] = (K_sun * ang_fac) / torch.abs(d_lnne).clamp_min(1e-6)
gamma_map = gamma_map.clamp(max=1e8)

gamma_np   = to_numpy(gamma_map)
E_grid_np  = to_numpy(ENERGY_GRID)
rho_np_g   = to_numpy(RHO_GRID)

fig, ax = plt.subplots(figsize=(9.5, 5.5))
import matplotlib.colors as mcolors
mesh = ax.pcolormesh(rho_np_g, E_grid_np, np.log10(gamma_np.clip(1e-2, 1e8)),
                     shading='auto', cmap='YlOrRd_r')
cbar = fig.colorbar(mesh, ax=ax, label=r'$\log_{10}\,\gamma(\rho, E)$')
# Contour at gamma = 1  (log10 = 0)
try:
    ax.contour(rho_np_g, E_grid_np, np.log10(gamma_np.clip(1e-2, 1e8)),
               levels=[0.0], colors='blue', linewidths=1.8, linestyles='--')
except Exception:
    pass
ax.set_xlabel(r'Solar radius $\rho = r/R_\odot$')
ax.set_ylabel('Energy $E$ [MeV]')
ax.set_title(r'Adiabaticity map $\log_{10}\gamma(\rho, E)$ — blue dashed: $\gamma=1$ boundary')
fig.tight_layout()
save_and_show('sn3_fig4_adiabaticity_map.png', fig, output_dir=OUTPUT_DIR, show_plots=config.show_plots)

idx10 = int(torch.argmin(torch.abs(ENERGY_GRID - 10.0)).item())
idx01 = int(torch.argmin(torch.abs(RHO_GRID - 0.1)).item())
print(f'gamma at (rho=0.1, E=10 MeV) = {gamma_np[idx10, idx01]:.1f}')
```

## 5. Adiabaticity Profile $\gamma(\rho)$ at Fixed Energy

$\gamma(\rho)$ is plotted as a function of solar radius at each probe energy. This cross section of the map above shows how adiabaticity varies **across the solar production region** for a given source energy.

**Expected results:**
- all curves are orders of magnitude above 1 across the entire solar interior for LMA parameters, confirming that the adiabatic approximation is robust.
- Curves decrease at larger $\rho$ where the density gradient is smaller and $K_\odot$ may become large.

```python
fig, ax = plt.subplots(figsize=(9, 5.2))
colors = plt.cm.plasma(np.linspace(0.15, 0.85, len(PROBE_ENERGIES)))
for E_val, col in zip(PROBE_ENERGIES, colors):
    K_sun  = kinetic_potential(osc_lma.mass_spectrum.DeltamSq21,
                               torch.tensor(E_val, dtype=ctx.dtype),
                               evolution_scale_m=constant.R_SUN)
    gamma_r = (K_sun * ang_fac) / torch.abs(d_lnne).clamp_min(1e-6)
    gamma_r = gamma_r.clamp(max=1e8)
    ax.semilogy(to_numpy(RHO_GRID), to_numpy(gamma_r), color=col, label=f'E={E_val:.1f} MeV')

ax.axhline(1.0, color='red', ls='--', lw=1.4, label=r'$\gamma=1$ (non-adiabatic onset)')
ax.set_xlabel(r'Solar radius $\rho = r/R_\odot$')
ax.set_ylabel(r'Adiabaticity parameter $\gamma(\rho)$')
ax.set_title('Adiabaticity across the solar production region (LMA parameters)')
ax.set_xlim(0, 1)
ax.legend(fontsize=9)
fig.tight_layout()
save_and_show('sn3_fig5_adiabaticity_profile.png', fig, output_dir=OUTPUT_DIR, show_plots=config.show_plots)
```

## 6. Adiabaticity Parameter vs Energy $\gamma_{\rm res}(E)$

$\gamma_{\rm res}(E)$ is evaluated **at the resonance radius** — the physically relevant quantity for the Landau-Zener transition probability (Notebook 3). Results are shown for the three oscillation scenarios.

**Expected results:**
- **LMA**: $\gamma_{\rm res}\sim 10^3$–$10^4$; at 10 MeV $\gamma\approx 1400$ — propagation extremely adiabatic.
- **Hyp** (÷100): $\gamma\sim 10$–100 — still well adiabatic.
- **N.A.** (÷1000): $\gamma$ approaches 1 at several MeV — Landau-Zener transitions become significant.

```python
E_g   = torch.logspace(-1, np.log10(25), 400, dtype=ctx.dtype)
E_g_np = to_numpy(E_g)

gamma_lma = to_numpy(_gamma_res_grid(osc_lma,      E_g, profile))
gamma_hyp = to_numpy(_gamma_res_grid(osc_hyp,      E_g, profile))
gamma_na  = to_numpy(_gamma_res_grid(osc_nonadiab, E_g, profile))

fig, ax = plt.subplots(figsize=(9, 5.2))
ax.semilogy(E_g_np, gamma_lma, '#1f77b4', lw=2.5, label=r'LMA — $\Delta m^2_{21}$ (NuFIT 5.2 NO)')
ax.semilogy(E_g_np, gamma_hyp, '#ff7f0e', lw=2.0, ls='--', label=r'Hyp — $\Delta m^2_{21}/100$')
ax.semilogy(E_g_np, gamma_na,  '#2ca02c', lw=2.0, ls=':', label=r'N.A. — $\Delta m^2_{21}/1000$')
ax.axhline(1.0, color='red', ls='--', lw=1.4, label=r'$\gamma_{\rm res}=1$')
ax.set_xlabel('$E$ [MeV]')
ax.set_ylabel(r'$\gamma_{\rm res}(E)$')
ax.set_xlim(0.1, 25)
ax.set_ylim(1e-1, 1e5)
ax.set_title(r'Adiabaticity parameter $\gamma_{\rm res}(E)$ evaluated at the resonance')
ax.legend(fontsize=9)
fig.tight_layout()
save_and_show('sn3_fig6_gamma_vs_energy.png', fig, output_dir=OUTPUT_DIR, show_plots=config.show_plots)

print(f'{"E [MeV]":>10} | {"γ_LMA":>14} | {"γ_Hyp":>14} | {"γ_NA":>14}')
print('-'*60)
for E_val in [1.0, 5.0, 10.0]:
    i = int(np.argmin(np.abs(E_g_np - E_val)))
    g_l = gamma_lma[i] if np.isfinite(gamma_lma[i]) else float('nan')
    g_h = gamma_hyp[i] if np.isfinite(gamma_hyp[i]) else float('nan')
    g_n = gamma_na[i]  if np.isfinite(gamma_na[i])  else float('nan')
    print(f'{E_val:>10.1f} | {g_l:>14.1f} | {g_h:>14.2f} | {g_n:>14.3f}')
```

## 7. Critical Energy and Solar Source Bands

The **critical energy** $E_{\rm crit}$ is the energy at which $\gamma_{\rm res}(E_{\rm crit}) = 1$, marking the onset of significant Landau-Zener corrections. The plot below overlays the $\gamma_{\rm res}(E)$ curves with the energy ranges of the main solar neutrino sources, making the connection to observable physics explicit.

**Expected results:** 
- For LMA, $E_{\rm crit}$ is far below 0.1 MeV (off the plot to the left) — all solar sources lie safely in the adiabatic regime.
- For the N.A. scenario, $E_{\rm crit}$ falls within the $^8$B energy window, meaning a visible LZ correction would be expected.

```python
src_bands = [
    ('pp',          0.10,  0.42, '#CCEE88', 0.55),
    (r'$^7$Be',     0.38,  0.90, '#88CCEE', 0.55),
    (r'pep',        1.00,  1.50, '#FFAA88', 0.65),
    (r'$^8$B',      1.00, 16.00, '#FFCC88', 0.40),
    ('hep',        14.00, 20.00, '#CC99EE', 0.55),
]

fig, ax = plt.subplots(figsize=(11, 5.5))

for lbl, emin, emax, col, alp in src_bands:
    ax.axvspan(emin, emax, alpha=alp*0.18, color=col, zorder=0)
    ax.text((emin+emax)/2, 5e4, lbl, ha='center', va='bottom', fontsize=8, color='0.4')

ax.semilogy(E_g_np, gamma_lma, '#1f77b4', lw=2.5, label=r'LMA — $\Delta m^2_{21}$ (NuFIT 5.2 NO)')
ax.semilogy(E_g_np, gamma_hyp, '#ff7f0e', lw=2.0, ls='--', label=r'Hyp — $\Delta m^2_{21}/100$')
ax.semilogy(E_g_np, gamma_na,  '#2ca02c', lw=2.0, ls=':',  label=r'N.A. — $\Delta m^2_{21}/1000$')
ax.axhline(1.0, color='red', ls='--', lw=1.5, label=r'$\gamma_{\rm res}=1$ (LZ onset)')

# Find and mark E_crit for each scenario
for lbl, gamma_arr, col in [('LMA', gamma_lma, '#1f77b4'),
                              ('Hyp', gamma_hyp, '#ff7f0e'),
                              ('N.A.', gamma_na, '#2ca02c')]:
    finite_mask = np.isfinite(gamma_arr) & (gamma_arr > 0)
    if finite_mask.any():
        sign = np.sign(gamma_arr[finite_mask] - 1.0)
        cross = np.where((sign[:-1] > 0) & (sign[1:] <= 0))[0]
        if len(cross) > 0:
            i0 = np.where(finite_mask)[0][cross[0]]
            E_crit = E_g_np[i0]
            ax.axvline(E_crit, color=col, lw=1.2, ls=':', alpha=0.8)
            ax.text(E_crit+0.65, 3, f'{lbl}\n$E_{{\\rm crit}}$={E_crit:.2f}', color=col,
                    fontsize=8, ha='center', va='bottom')
        else:
            print(f'{lbl}: E_crit below plot range (gamma always > 1)')

ax.set_xlabel('$E$ [MeV]')
ax.set_ylabel(r'$\gamma_{\rm res}(E)$')
ax.set_xlim(0.1, 20)
ax.set_ylim(1e-1, 1e5)
ax.set_title(r'$\gamma_{\rm res}(E)$ with solar source bands — $E_{\rm crit}$ where $\gamma=1$')
ax.legend(fontsize=8)
fig.tight_layout()
save_and_show('sn3_fig7_gamma_sources.png', fig, output_dir=OUTPUT_DIR, show_plots=config.show_plots)
```

## 8. Summary

| Quantity | LMA | Hyp (÷100) | N.A. (÷1000) |
|----------|-----|------------|---------------|
| $\gamma_{\rm res}$ at 1 MeV | No Resonance | $\sim \mathcal{O}^2$ | $\sim \mathcal{O}^1$ |
| $\gamma_{\rm res}$ at 10 MeV | $\sim \mathcal{O}^3$ | $\sim \mathcal{O}^1$ | $\sim \mathcal{O}^0$ |
| $E_{\rm crit}$ | $\ll 0.1$ MeV (off-scale) | $\sim 1$ MeV | $\sim 10$ MeV |
| pp source adiabatic | ✓ trivially | ✓ | ✓ (approximated) |
| $^8$B source adiabatic | ✓ exact | ✓ excellent | ✗ LZ needed |



**Physical interpretation:**

1. For standard LMA parameters, $\gamma_{\rm res}\gg 1$ across the entire solar-neutrino energy range. The adiabatic approximation is not merely adequate — it is **numerically exact** (see Notebook over LZ Correction for the quantitative proof via $P_{\rm LZ}\equiv 0$).
2. The key quantity determining adiabaticity is the **density scale height at the resonance**, $L_n(r_{\rm res})$. In the LMA case the resonance sits in the dense solar core ($r_{\rm res}\lesssim 0.3\,R_\odot$) where $L_n$ is large relative to the oscillation length — precisely the combination that yields large $\gamma$.
3. Hypothetical scenarios with smaller $\Delta m^2_{21}$ push the resonance into the outer envelope where $L_n$ is smaller, partially compensating the naive $\gamma\propto\Delta m^2_{21}$ scaling. The actual reduction in $\gamma$ is therefore less than the ratio of the splittings.
4. The observable consequence of non-adiabaticity is a Landau-Zener jump between matter eigenstates, computed in Next Notebook.
