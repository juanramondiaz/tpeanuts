# PREM radial Earth model — EDA
Direct source table associated with Dziewonski & Anderson (1981). PREM constrains mass density and seismic properties; it does not provide electron fraction $Y_e$.

```python
from pathlib import Path
import pandas as pd, matplotlib.pyplot as plt
ROOT=Path.cwd().resolve().parents[3]
path=ROOT/'data/earth/prem/raw/prem_table_full.txt'
rows=[]
for line in path.read_text().splitlines()[2:]:
    fields=line.split()
    if len(fields)==12:
        try: rows.append([float(value) if value != '???' else float('nan') for value in fields])
        except ValueError: pass
table=pd.DataFrame(rows,columns=['radius_km','rho','vp','vs','phi','K','lambda','mu','nu','g','P','depth_km'])
display(table.describe())
table.plot(x='radius_km',y='rho',drawstyle='steps-post',title='PREM mass density'); plt.show()
```
