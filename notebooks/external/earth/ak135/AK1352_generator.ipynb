# ak135-f canonical generator
Preserves seismic columns and derives neutrino composition columns using the same documented $Y_e$ assumptions as PREM.

```python
from pathlib import Path
import pandas as pd
ROOT=Path.cwd().resolve().parents[3]; raw=ROOT/'data/earth/ak135/raw/AK135F_AVG.csv'
t=pd.read_csv(raw,header=None,names=['depth_km','mass_density_g_cm3','vp_km_s','vs_km_s','q_alpha','q_mu'])
t['radius_km']=6371.0-t.depth_km; t['radius_fraction']=t.radius_km/6371.0
t['electron_fraction']=t.depth_km.map(lambda d: .5556 if d<3 else (.467 if d>2891 else .495))
t['electron_density_mol_cm3']=t.mass_density_g_cm3*t.electron_fraction
t['neutron_density_mol_cm3']=t.mass_density_g_cm3*(1-t.electron_fraction)
out=ROOT/'data/earth/ak135/density/ak135f_density.csv'; t.sort_values('radius_km').to_csv(out,index=False)
print(out,t.shape); display(t.head())
```
