# ak135-f radial Earth model — EDA
The raw CSV comes directly from the NSF SAGE/EarthScope EMC product. Density is in Mg m$^{-3}$, numerically equal to g cm$^{-3}$.

```python
from pathlib import Path
import pandas as pd, matplotlib.pyplot as plt
ROOT=Path.cwd().resolve().parents[3]
path=ROOT/'data/earth/ak135/raw/AK135F_AVG.csv'
t=pd.read_csv(path,header=None,names=['depth_km','rho','vp','vs','q_alpha','q_mu'])
display(t.describe()); t.plot(x='depth_km',y='rho',drawstyle='steps-post',title='ak135-f mass density'); plt.gca().invert_xaxis(); plt.show()
```
