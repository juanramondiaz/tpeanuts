# Bartol atmospheric-neutrino data — EDA
Reference: Barr et al., Phys. Rev. D 70, 023006 (2004). The historical Oxford download endpoint is recorded in `data/atmosphere/bartol/metadata/source.json` but is currently unavailable.

```python
from pathlib import Path
import json
ROOT=Path.cwd().resolve().parents[3]
META=ROOT/'data/atmosphere/bartol/metadata/source.json'
metadata=json.loads(META.read_text())
display(metadata)
print('Raw files:', list((META.parents[1]/'raw').glob('*')))
```

## Important unit distinction
Bartol tables publish $dN/d\ln E$. Conversion to the canonical differential-energy convention requires $dN/dE=(1/E)dN/d\ln E$. The generator makes that conversion explicitly.
