# Bartol canonical generator
Place an original five-column BGLRS table in `data/atmosphere/bartol/raw`. This notebook refuses to substitute another provider when Bartol data are absent.

```python
from pathlib import Path
import pandas as pd
ROOT=Path.cwd().resolve().parents[3]
RAW_DIR=ROOT/'data/atmosphere/bartol/raw'
files=sorted(RAW_DIR.glob('*.dat'))+sorted(RAW_DIR.glob('*.txt'))
if not files:
    raise FileNotFoundError('No original Bartol table found; see metadata/source.json')
frames=[]
for path in files:
    raw=pd.read_csv(path,sep=r'\s+',comment='#',header=None,names=['energy_GeV','cos_zenith','dN_dlnE','mc_error','events'])
    raw['flux']=raw.dN_dlnE/raw.energy_GeV
    raw['particle']=path.stem.split('_')[-1]
    frames.append(raw[['energy_GeV','cos_zenith','particle','flux']])
table=pd.concat(frames,ignore_index=True)
out=ROOT/'data/atmosphere/bartol/flux/bartol_flux.csv'; table.to_csv(out,index=False)
print(out,table.shape)
```
